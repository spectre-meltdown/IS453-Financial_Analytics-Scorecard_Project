## 1.3 Bureau Data Aggregation & Left Merge

```python
# ── Aggregate bureau data to one row per applicant ──

bureau_num_cols = bureau_data.select_dtypes(include='number').columns \
    .drop(['SK_ID_CURR', 'SK_ID_BUREAU']).tolist()

# Numerical aggregations
bureau_num_agg = bureau_data.groupby('SK_ID_CURR')[bureau_num_cols] \
    .agg(['mean', 'max', 'min', 'sum'])
bureau_num_agg.columns = [
    f'BUREAU_{col}_{agg.upper()}' for col, agg in bureau_num_agg.columns]
bureau_num_agg = bureau_num_agg.reset_index()

# Record count
bureau_count = bureau_data.groupby('SK_ID_CURR').size() \
    .reset_index(name='BUREAU_RECORD_COUNT')

# Credit-active status pivoted to counts
credit_active_pivot = bureau_data.groupby('SK_ID_CURR')['CREDIT_ACTIVE'] \
    .value_counts().unstack(fill_value=0)
credit_active_pivot.columns = [
    f'BUREAU_ACTIVE_{c.replace(" ", "_").upper()}'
    for c in credit_active_pivot.columns]
credit_active_pivot = credit_active_pivot.reset_index()

# Distinct credit-type count
credit_type_cnt = bureau_data.groupby('SK_ID_CURR')['CREDIT_TYPE'] \
    .nunique().reset_index(name='BUREAU_CREDIT_TYPE_COUNT')

# Car-loan history flag
car_ids = bureau_data.loc[
    bureau_data['CREDIT_TYPE'] == 'Car loan', 'SK_ID_CURR'].unique()
car_loan_flag = pd.DataFrame({
    'SK_ID_CURR': car_ids, 'HAS_CAR_LOAN_HISTORY': 1})

# Combine all bureau features
bureau_agg = bureau_count \
    .merge(bureau_num_agg, on='SK_ID_CURR', how='left') \
    .merge(credit_active_pivot, on='SK_ID_CURR', how='left') \
    .merge(credit_type_cnt, on='SK_ID_CURR', how='left') \
    .merge(car_loan_flag, on='SK_ID_CURR', how='left')

bureau_agg['HAS_CAR_LOAN_HISTORY'] = bureau_agg['HAS_CAR_LOAN_HISTORY'].fillna(0).astype(int)

print(f"Bureau aggregated: {bureau_agg.shape}")
print(f"Columns: {bureau_agg.columns.tolist()}")

# ── Left merge ──
merged_df = application_data.merge(bureau_agg, on='SK_ID_CURR', how='left')

# Derived fields
merged_df['AGE_YEARS']  = (-merged_df['DAYS_BIRTH'] / 365.25).round(1)
merged_df['EMP_YEARS']  = (
    merged_df['DAYS_EMPLOYED'].replace(365243, np.nan).abs() / 365.25
).round(1)
merged_df['HAS_CAR_LOAN_HISTORY'] = merged_df['HAS_CAR_LOAN_HISTORY'].fillna(0).astype(int)

print(f"\nMerged DataFrame: {merged_df.shape}")
print(f"With bureau data:    {merged_df['BUREAU_RECORD_COUNT'].notna().sum():,}")
print(f"Without bureau data: {merged_df['BUREAU_RECORD_COUNT'].isna().sum():,}")
```

```python
# Apply GROUP 5 filter on fully merged data
group5_filter = (
    (merged_df['NAME_CONTRACT_TYPE'] == 'Cash loans') &
    (merged_df['FLAG_OWN_CAR'] != 'Y') &
    (merged_df['HAS_CAR_LOAN_HISTORY'] != 1) &
    (merged_df['AGE_YEARS'].between(21, 65))
)
group5_df = merged_df[group5_filter].copy()

print(f"GROUP 5 rows: {len(group5_df):,}  "
      f"({len(group5_df)/len(merged_df)*100:.1f}% of total)")
print(f"GROUP 5 bad rate: {group5_df['TARGET'].mean():.4%}")
print(f"Overall bad rate:  {merged_df['TARGET'].mean():.4%}")
print(f"Shape: {group5_df.shape}")
```

## 1.5 Outlier Detection

```python
# Outlier detection — IQR method
num_cols_eda = group5_df.select_dtypes(include='number').columns.tolist()
num_cols_eda = [c for c in num_cols_eda if c not in ['SK_ID_CURR', 'TARGET']]

outlier_rows = []
for col in num_cols_eda:
    data = group5_df[col].dropna()
    if len(data) == 0:
        continue
    Q1, Q3 = data.quantile(0.25), data.quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    n_out = ((data < lower) | (data > upper)).sum()
    outlier_rows.append({
        'column': col,
        'total': len(data),
        'outlier_count': n_out,
        'outlier_pct': round(n_out / len(data) * 100, 2),
        'lower_bound': round(lower, 2),
        'upper_bound': round(upper, 2),
        'min': round(data.min(), 2),
        'max': round(data.max(), 2),
    })

outlier_df = pd.DataFrame(outlier_rows).sort_values('outlier_pct', ascending=False)
print(f"Columns with outliers: "
      f"{(outlier_df['outlier_count'] > 0).sum()} / {len(outlier_df)}")
display(outlier_df.head(30))

# Box-plots for top-6 outlier columns
top6 = outlier_df.head(6)['column'].tolist()
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
for i, col in enumerate(top6):
    ax = axes[i // 3, i % 3]
    group5_df[col].dropna().plot(kind='box', ax=ax)
    pct = outlier_df.loc[outlier_df['column'] == col, 'outlier_pct'].values[0]
    ax.set_title(f'{col}\n({pct}% outliers)')
plt.suptitle('Top 6 Columns by Outlier %', y=1.02)
plt.tight_layout()
plt.show()

print("\n=== PROPOSED OUTLIER TREATMENT ===")
print("Valid outliers (domain-expected extreme values):")
print("  - AMT_INCOME_TOTAL: high incomes are valid → cap at 99th pctile")
print("  - DAYS_EMPLOYED: 365243 placeholder already replaced with NaN")
print("  - CNT_CHILDREN: large families valid → cap at 5+")
print("Invalid outliers (likely data errors):")
print("  - Negative values in amount fields → flag for review")
print("  - Illogical DAYS values → review and treat")
```

## 1.6 Missing Data Analysis & Treatment

```python
# Missing-data analysis on GROUP 5 merged data
missing_g5 = pd.DataFrame({
    'null_count': group5_df.isnull().sum(),
    'null_pct': (group5_df.isnull().sum() / len(group5_df) * 100).round(2),
    'dtype': group5_df.dtypes
}).sort_values('null_pct', ascending=False)
missing_g5 = missing_g5[missing_g5['null_pct'] > 0]

high_missing   = missing_g5[missing_g5['null_pct'] > 40]
medium_missing = missing_g5[(missing_g5['null_pct'] > 5) &
                            (missing_g5['null_pct'] <= 40)]
low_missing    = missing_g5[missing_g5['null_pct'] <= 5]

print(f"Columns with missing values: {len(missing_g5)} / {len(group5_df.columns)}")

print(f"\nHigh missing (>40%) — {len(high_missing)} cols → DROP")
display(high_missing)
print(f"\nMedium missing (5–40%) — {len(medium_missing)} cols → IMPUTE + FLAG")
display(medium_missing)
print(f"\nLow missing (≤5%) — {len(low_missing)} cols → IMPUTE")
display(low_missing)

# Visualisation
fig, ax = plt.subplots(figsize=(16, 6))
missing_g5['null_pct'].head(40).plot(kind='bar', ax=ax, color='coral')
ax.axhline(y=40, color='red', linestyle='--', label='40% threshold (drop)')
ax.axhline(y=5,  color='orange', linestyle='--', label='5% threshold')
ax.set_title('Missing Values — GROUP 5 Merged Data')
ax.set_ylabel('Missing %')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# ── Missing-data treatment ──

# Columns to drop (>40% missing)
cols_to_drop_missing = high_missing.index.tolist()

# Bureau columns: NaN → 0 (no bureau record means zero)
bureau_fill_cols = [c for c in group5_df.columns if c.startswith('BUREAU_')]
group5_df[bureau_fill_cols] = group5_df[bureau_fill_cols].fillna(0)

# Low-missing numerical → median
num_low = [c for c in low_missing.index
           if low_missing.loc[c, 'dtype'] != 'object'
           and c not in bureau_fill_cols]
for col in num_low:
    group5_df[col] = group5_df[col].fillna(group5_df[col].median())

# Low-missing categorical → mode
cat_low = [c for c in low_missing.index
           if low_missing.loc[c, 'dtype'] == 'object']
for col in cat_low:
    group5_df[col] = group5_df[col].fillna(group5_df[col].mode()[0])

# Medium-missing → missing-indicator flag + impute
for col in medium_missing.index:
    if col in bureau_fill_cols:
        continue
    group5_df[f'{col}_MISSING'] = group5_df[col].isnull().astype(int)
    if group5_df[col].dtype == 'object':
        group5_df[col] = group5_df[col].fillna('MISSING')
    else:
        group5_df[col] = group5_df[col].fillna(group5_df[col].median())

# Drop high-missing columns
group5_df = group5_df.drop(columns=cols_to_drop_missing, errors='ignore')

print(f"After treatment:")
print(f"  Dropped {len(cols_to_drop_missing)} columns (>40% missing)")
print(f"  Shape:  {group5_df.shape}")
print(f"  Remaining nulls: {group5_df.isnull().sum().sum()}")
```

## 1.7 VIF — Multicollinearity Check

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Select numerical columns for VIF
vif_candidates = group5_df.select_dtypes(include='number').columns.tolist()
vif_candidates = [c for c in vif_candidates
                  if c not in ['SK_ID_CURR', 'TARGET']
                  and group5_df[c].nunique() > 1]

# Use top-30 by TARGET correlation (VIF is O(p²n))
target_corr_abs = group5_df[vif_candidates + ['TARGET']].corr()['TARGET'] \
    .drop('TARGET').abs().sort_values(ascending=False)
vif_candidates = target_corr_abs.head(30).index.tolist()
print(f"VIF on top 30 features (by |corr with TARGET|)")

# Sample for speed & standardise
sample = group5_df[vif_candidates].sample(
    min(10_000, len(group5_df)), random_state=42).dropna()
sample_std = (sample - sample.mean()) / sample.std()
sample_std = sample_std.dropna(axis=1)           # drop any all-NaN cols

vif_results = pd.DataFrame({
    'feature': sample_std.columns,
    'VIF': [variance_inflation_factor(sample_std.values, i)
            for i in range(sample_std.shape[1])]
}).sort_values('VIF', ascending=False)

print("\nVIF  (>5 = moderate,  >10 = severe multicollinearity)")
display(vif_results)

high_vif = vif_results[vif_results['VIF'] > 5]
print(f"\nFeatures with VIF > 5: {len(high_vif)}")
if len(high_vif) > 0:
    display(high_vif)
```

## 2.1 WoE / IV — Fine Classing

```python
def calculate_woe_iv(df, feature, target, bins=10):
    """Calculate Weight of Evidence and Information Value."""
    d = df[[feature, target]].copy().dropna()

    # Binning
    if d[feature].dtype == 'object' or d[feature].nunique() <= bins:
        d['bin'] = d[feature]
    else:
        d['bin'] = pd.qcut(d[feature], q=bins, duplicates='drop')

    grouped = d.groupby('bin', observed=True)[target].agg(['count', 'sum'])
    grouped.columns = ['total', 'bad']
    grouped['good'] = grouped['total'] - grouped['bad']

    total_bad  = grouped['bad'].sum()
    total_good = grouped['good'].sum()

    grouped['bad_dist']  = grouped['bad']  / total_bad
    grouped['good_dist'] = grouped['good'] / total_good

    # Avoid log(0)
    grouped['bad_dist']  = grouped['bad_dist'].replace(0, 0.0001)
    grouped['good_dist'] = grouped['good_dist'].replace(0, 0.0001)

    grouped['woe'] = np.log(grouped['good_dist'] / grouped['bad_dist'])
    grouped['iv']  = (grouped['good_dist'] - grouped['bad_dist']) * grouped['woe']

    return grouped, grouped['iv'].sum()


# ── Compute IV for every feature ──
target_col = 'TARGET'
feature_cols = [c for c in group5_df.columns
                if c not in ['SK_ID_CURR', 'TARGET']]

iv_results = []
woe_details = {}

for col in feature_cols:
    try:
        woe_tbl, iv = calculate_woe_iv(group5_df, col, target_col)
        iv_results.append({'feature': col, 'IV': round(iv, 4)})
        woe_details[col] = woe_tbl
    except Exception:
        iv_results.append({'feature': col, 'IV': 0})

iv_df = pd.DataFrame(iv_results).sort_values('IV', ascending=False)
iv_df['predictive_power'] = pd.cut(
    iv_df['IV'],
    bins=[-np.inf, 0.02, 0.1, 0.3, 0.5, np.inf],
    labels=['Not Predictive', 'Weak', 'Medium', 'Strong',
            'Suspicious (overfit)'])

print("IV interpretation:")
print("  < 0.02  Not predictive")
print("  0.02–0.10  Weak")
print("  0.10–0.30  Medium")
print("  0.30–0.50  Strong")
print("  > 0.50  Suspicious (potential overfitting)\n")
display(iv_df.head(40))
```

## 2.2 Coarse Classing

```python
def coarse_class(df, feature, target, max_bins=5):
    """Reduce bins to achieve near-monotonic WoE trend."""
    d = df[[feature, target]].copy().dropna()

    if d[feature].dtype == 'object':
        # Group small categories into OTHER
        vc = d[feature].value_counts(normalize=True)
        small = vc[vc < 0.05].index.tolist()
        if small:
            d.loc[d[feature].isin(small), feature] = 'OTHER'
        return calculate_woe_iv(d, feature, target, bins=max_bins)

    # Try bin counts from max_bins down to 3; pick most monotonic
    best, best_score = None, -1
    for nb in range(max_bins, 2, -1):
        try:
            result, iv = calculate_woe_iv(d, feature, target, bins=nb)
            woe_vals = result['woe'].values
            if len(woe_vals) > 1:
                diffs = np.diff(woe_vals)
                mono = max(np.mean(diffs >= 0), np.mean(diffs <= 0))
            else:
                mono = 1.0
            if mono > best_score:
                best_score = mono
                best = (result, iv)
        except Exception:
            continue
    return best if best else calculate_woe_iv(d, feature, target, bins=3)


# Apply coarse classing to top-IV features
coarse_feats = iv_df[iv_df['IV'] >= 0.02].head(15)['feature'].tolist()
coarse_results = {}
for feat in coarse_feats:
    try:
        tbl, iv = coarse_class(group5_df, feat, target_col)
        coarse_results[feat] = {'woe_table': tbl, 'iv': iv}
    except Exception as e:
        print(f"Skip {feat}: {e}")

# Visualise
n = len(coarse_results)
ncols = 3
nrows = (n + ncols - 1) // ncols
fig, axes = plt.subplots(nrows, ncols, figsize=(18, 4 * nrows))
axes_flat = axes.flatten()

for i, (feat, res) in enumerate(coarse_results.items()):
    tbl = res['woe_table']
    ax = axes_flat[i]
    woe_vals = tbl['woe'].values
    labels = [str(x)[:25] for x in tbl.index]
    bar_colors = ['steelblue' if v >= 0 else 'coral' for v in woe_vals]
    ax.bar(range(len(woe_vals)), woe_vals, color=bar_colors)
    ax.set_title(f'{feat}\nIV={res["iv"]:.4f}  ({len(woe_vals)} bins)',
                 fontsize=9)
    ax.set_xticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=7)
    ax.axhline(y=0, color='black', lw=0.5)
    ax.set_ylabel('WoE')

for j in range(i + 1, len(axes_flat)):
    axes_flat[j].set_visible(False)

plt.suptitle('Coarse Classing — WoE by Bin', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```
